Create a Histogram of Salary Feature

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Salary', color = 'orange',
             edgecolor = 'linen', alpha = 0.7, bins = 10)

plt.title("Histogram of Salary")
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

From the Histogram we can observe the distribution of salaries in the dataset. The Histogram shows that majority of employees are in lower salaries range and as the salary increases the count of employees decreases significantly.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Salary', y='Age')
plt.title('Scatter Plot of Employee Salary vs Employee Age')
plt.xlabel('Salary')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

This scatter plot shows a general upward trend in the scatter plot. This shows that as age increases there is a tendency for salary also to increase. There are areas where there are more data points such as in lower salary ranges. It shows that age is a factor influencing salary with older employees generally earning

more. However salary is not solely determined by age. Other factors also contribute to the variability in salaries.

```python
# Convert 'Attrition' to numerical (1 for Yes, 0 for No)
data['Attrition_numeric'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate the mean attrition rate for each job satisfaction level
attrition_rate_by_job_satisfaction = data.groupby('JobSatisfaction')['Attrition_numeric'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=attrition_rate_by_job_satisfaction, x='JobSatisfaction', y='Attrition_numeric', color='teal')
plt.title('Attrition Rate by Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Rate')
plt.ylim(0, 1) # Attrition rate is between 0 and 1
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x='JobSatisfaction', y='Attrition', ci=None)
plt.title('Attrition Rate by Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Rate')
plt.show()
```

Here are some different types of plots available in the `seaborn` library:

*   **Relational plots:**
    *   `scatterplot()`: Scatter plot
    *   `lineplot()`: Line plot

*   **Distribution plots:**
    *   `histplot()`: Histogram
    *   `kdeplot()`: Kernel Density Estimate plot
    *   `ecdfplot()`: Empirical Cumulative Distribution Function plot
    *   `rugplot()`: Adds ticks on the x-axis representing the location of each observation
    *   `displot()`: Figure-level function for plotting distributions

*   **Categorical plots:**
    *   `stripplot()`: Scatter plot where one variable is categorical
    *   `swarmplot()`: Similar to stripplot, but points are adjusted to avoid overlap
    *   `boxplot()`: Box plot
    *   `violinplot()`: Violin plot (combination of box plot and KDE plot)
    *   `pointplot()`: Plots the mean of the quantitative variable for each category
    *   `barplot()`: Bar plot
    *   `countplot()`: Shows the counts of observations in each categorical bin
    *   `catplot()`: Figure-level function for plotting categorical data

*   **Regression plots:**
    *   `regplot()`: Scatter plot with a linear regression fit
    *   `lmplot()`: Figure-level function for plotting regression models

*   **Matrix plots:**
    *   `heatmap()`: Heatmap of a matrix
    *   `clustermap()`: Hierarchically-clustered heatmap

*   **Multi-variate plots:**
    *   `pairplot()`: Pairwise relationships in a dataset
    *   `jointplot()`: Joint distribution of two variables

*   **Grids:**
    *   `FacetGrid`: Create a grid of plots based on categorical variables
    *   `JointGrid`: Create a grid for bivariate plots
