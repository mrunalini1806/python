```python
import numpy as np
import pandas as pd
%pip install seaborn
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline
sns.set(color_codes=True)
from scipy import stats
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('greendestination (1).csv')
```

```python
# Analyzing data;
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
#Deleting Unnecessary Columns
data.drop(['DailyRate','HourlyRate','MonthlyRate','MaritalStatus','StockOptionLevel','DistanceFromHome'],axis=1,inplace=True)
```

```python
data.columns
```

```python
#Renaming Columns
data.rename(columns={'MonthlyIncome':'Salary'},inplace=True)
```

```python
data.info()
```

Create a Histogram of Salary Feature

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Salary', color = 'orange',
             edgecolor = 'linen', alpha = 0.7, bins = 10)

plt.title("Histogram of Salary")
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

From the Histogram we can observe the distribution of salaries in the dataset. The Histogram shows that majority of employees are in lower salaries range and as the salary increases the count of employees decreases significantly.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Salary', y='Age')
plt.title('Scatter Plot of Employee Salary vs Employee Age')
plt.xlabel('Salary')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

This scatter plot shows a general upward trend in the scatter plot. This shows that as age increases there is a tendency for salary also to increase. There are areas where there are more data points such as in lower salary ranges. It shows that age is a factor influencing salary with older employees generally earning

more. However salary is not solely determined by age. Other factors also contribute to the variability in salaries.

```python
average_job_satisfaction_by_role = data.groupby('JobRole')['JobSatisfaction'].mean().reset_index()
print(average_job_satisfaction_by_role)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='JobRole', hue='JobSatisfaction', palette='viridis')
plt.title('Job Satisfaction Distribution by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

We have created a count plot to analyze the relationship between JobRole and Job Saisfaction Levels across different job roles.This is a grouped bar plot showing the distribution of Job satisfaction levels within each job role. Each group of bars represents a specific Job role and the bars within each group show the count of employees at each Job satisfaction level for that role. While the Job roles like 'Healthcare Representative and 'Research Scientist' appear to have slightly higher Job satisfaction, 'Human Resources' has the lowest job satisfaction.

```python
# Filter data for Sales representative job role
sales_rep_data = data[data['JobRole'] == 'Sales Representative'].copy()

# Plot the distribution of EnvironmentSatisfaction for Sales Representative
plt.figure(figsize=(6, 4))
sns.countplot(data=sales_rep_data, x='EnvironmentSatisfaction', palette='viridis')
plt.title('Environment Satisfaction Distribution for Sales Representative')
plt.xlabel('Environment Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

```python
# Filter data for Human Resources job role
hr_data = data[data['JobRole'] == 'Human Resources'].copy()

# Plot the distribution of EnvironmentSatisfaction for Human Resources
plt.figure(figsize=(6, 4))
sns.countplot(data=hr_data, x='EnvironmentSatisfaction', palette='viridis')
plt.title('Environment Satisfaction Distribution for Human Resources')
plt.xlabel('Environment Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(data=hr_data, x='JobSatisfaction', y='Salary', jitter=True, alpha=0.7, palette='viridis')
plt.title('Salary vs Job Satisfaction for Human Resources')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Salary')
plt.grid(axis='y')
plt.show()
```

Here the stipplot is showing the relationship between salary and job satisfaction for employees in Human resources.While there are some with higher salaries at higher job satisfaction levels there are also employees with relatively higher salaries at lower job satisfaction levels and vice versa. This shows that salary is not a strong predictor of job satisfaction.

Above is a count plot showing environment satisfaction among Human Resources employees. The highest number of employees reporting a satisfaction level of 3

```python
# Convert 'Attrition' to numerical (1 for Yes, 0 for No)
data['Attrition_numeric'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate the mean attrition rate for each job satisfaction level
attrition_rate_by_job_satisfaction = data.groupby('JobSatisfaction')['Attrition_numeric'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=attrition_rate_by_job_satisfaction, x='JobSatisfaction', y='Attrition_numeric', color='teal')
plt.title('Attrition Rate by Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Rate')
plt.ylim(0, 1) # Attrition rate is between 0 and 1
plt.show()
```

```python
# Calculate the mean attrition rate for each job satisfaction level in Human Resources
attrition_rate_by_job_satisfaction_hr = hr_data.groupby('JobSatisfaction')['Attrition_numeric'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=attrition_rate_by_job_satisfaction_hr, x='JobSatisfaction', y='Attrition_numeric', color='coral')
plt.title('Attrition Rate by Job Satisfaction Level for Human Resources')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Rate')
plt.ylim(0, 1) # Attrition rate is between 0 and 1
plt.show()
```

Above is the bar plot showing the relationship between job satisfaction and attrition in Human Resources Dept. This shows that the attrition is due to low job satisfaction.

```python
# Calculate the mean attrition rate for each relationship satisfaction level in Human Resources
attrition_rate_by_relationship_satisfaction_hr = hr_data.groupby('RelationshipSatisfaction')['Attrition_numeric'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=attrition_rate_by_relationship_satisfaction_hr, x='RelationshipSatisfaction', y='Attrition_numeric', color='purple')
plt.title('Attrition Rate by Relationship Satisfaction Level for Human Resources')
plt.xlabel('Relationship Satisfaction Level')
plt.ylabel('Attrition Rate')
plt.ylim(0, 1) # Attrition rate is between 0 and 1
plt.show()
```

Above is the bar plot showing relationship between attrition and relationship satisfaction. The attrition rate is highest for employees with relationship satisfaction of level 3. This indicates that other factors , in addition to relationship satisfaction are likely influencing attrition in the HR dept.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=hr_data, x='Attrition', y='PercentSalaryHike', palette='viridis')
plt.title('Percent Salary Hike vs Attrition for Human Resources')
plt.xlabel('Attrition')
plt.ylabel('Percent Salary Hike')
plt.show()
```

Above is the box plot visualizing the distribution of salary hikes for HR employees. It suggests employees who received a lower percentage salary hike might be somewhat more likely to leave. However there is an outlier in the 'Yes' group with higher salary hike.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=hr_data, x='PerformanceRating', hue='JobSatisfaction', palette='viridis')
plt.title('Job Satisfaction Distribution by Performance Rating for Human Resources')
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.show()
```

The above plot suggests in Human resources deparment, employees with a higher performance rating have higher job satisfaction.

```python
mean_standard_hours_hr = hr_data['StandardHours'].mean()
print(f"The mean standard work hours in Human Resources is: {mean_standard_hours_hr}")
```

```python
overtime_count_hr = hr_data[hr_data['OverTime'] == 'Yes'].shape[0]
total_employees_hr = hr_data.shape[0]
overtime_percentage_hr = (overtime_count_hr / total_employees_hr) * 100 if total_employees_hr > 0 else 0

print(f"Number of Human Resources employees working overtime: {overtime_count_hr}")
print(f"Percentage of Human Resources employees working overtime: {overtime_percentage_hr:.2f}%")
```

Abve is the count plot showing relationship between Overtime and Job Satisfaction. It shows that employees who work overtime might tend to have lower job satisfacton compared to those who do not work overtime.

 The overall analysis shows that there is an attrition particularly in Human resources department and the reasons for attrition are performance rating, overtime.
